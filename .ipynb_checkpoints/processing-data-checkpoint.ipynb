```python
import pandas as pd
import numpy as np
import os

import warnings
warnings.filterwarnings('ignore')
```

**Khong su dung index_col**

```python
path ='./data_for_example'

#Tạo một DataFrame rỗng
all_data = pd.DataFrame()

#Đọc dữ liệu trong từng file dữ liệu và tổng hợp thành một DataFrame chung
for file in os.listdir(path):
    path_file = path+'/'+file
    current_data = pd.read_csv(path_file)
    print(file, '---',current_data.shape[0])
    
    all_data = all_data.append(current_data)
```

```python
path ='./data_for_example'

#Tạo một DataFrame rỗng
all_data = pd.DataFrame()

#Đọc dữ liệu trong từng file dữ liệu và tổng hợp thành một DataFrame chung
for file in os.listdir(path):
    path_file = path+'/'+file
    current_data = pd.read_csv(path_file,index_col=0)
    print(file, '---',current_data.shape[0])
    
    all_data = all_data.append(current_data)
```

```python
all_data.head()
```

**Cột cụm thi không cần thiết lắm bởi vì 2 số đầu đã biểu diễn cụm thi rồi**

```python
all_data.drop(["Cụm thi"], axis=1, inplace=True)
```
